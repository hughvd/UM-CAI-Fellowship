## Counting words

```python
# Calculate the word count for each description
courses['word_count'] = courses['description'].apply(lambda x: len(x.split()))

# Group the DataFrame by department and calculate the mean word count
avg_word_count = courses.groupby('department')['word_count'].mean().reset_index()

# Rename the columns for clarity
avg_word_count.columns = ['department', 'avg_words_per_description']
```

```python
avg_word_count = avg_word_count.sort_values(by='avg_words_per_description', ascending=False)
```

```python
# Increase the maximum number of rows and columns to display
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
print(avg_word_count)
```
